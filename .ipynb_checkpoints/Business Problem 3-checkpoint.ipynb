#### Loss Ratio

##### Loss Ratio by Vehicle Class

In insurance industry, loss ratio is calculated either by annual claim amount divided by annual premium or total claim for the period divided by total premium for the period. <br>
In this dataset, we do not know the period in Total.Claim.Amount and the premium data is shown per month.
We assume that the Total.Claim.Amount is for the number of Months.Since.Policy.Inception and Total.Premium is calculated by multiplying Monthly.Premium.Auto with Months.Since.Policy.Inception

```r
cartypeclaim <- data %>%
    group_by(Vehicle.Class) %>%
    summarise(VehClass.Median.Claim = median(Total.Claim.Amount), 
              Total.VehClass.Median.Premium = median(Monthly.Premium.Auto*Months.Since.Policy.Inception)) %>%
    mutate(Margin = Total.VehClass.Median.Premium - VehClass.Median.Claim, 
           LossRatio = VehClass.Median.Claim/Total.VehClass.Median.Premium * 100)
cartypeclaim
```

Four-Door car and Two-Door Car seems to have better loss ratio than other class of vehicles

```r
data %>%
    group_by(Vehicle.Size) %>%
    summarise(VehClass.Median.Claim = median(Total.Claim.Amount), 
              Total.VehClass.Median.Premium = median(Monthly.Premium.Auto*Months.Since.Policy.Inception)) %>%
    mutate(Margin = Total.VehClass.Median.Premium - VehClass.Median.Claim, 
           LossRatio = VehClass.Median.Claim/Total.VehClass.Median.Premium * 100)
```

Small vehicle has higher loss ratio than Large and Medsize vehicles

#### 4. Bagged Tree on Balanced train_set without one hot encoding

If we want to estimate the model's accuracy using the "out-of-bag" (OOB) samples, we can set the the coob parameter to TRUE. The OOB samples are the training obsevations that were not selected into the bootstrapped sample (used in training). Since these observations were not used in training, we can use them instead to evaluate the accuracy of the model (done automatically inside the bagging() function).

```r
set.seed(123)
bagtree1 <- bagging(Response ~ ., 
                    data = train_set_n_bal,
                    method = 'class',
                    coob = TRUE)
```
